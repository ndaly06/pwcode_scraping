```python
#
import heapq
import nltk
nltk.download('averaged_perceptron_tagger')
import pandas as pd
import re
import requests
import json
import numpy as np
from tika import parser
```

```python
# bert paper
content = parser.from_file('https://arxiv.org/pdf/1810.04805v2.pdf')

raw_text = content['content']

text = str(raw_text)
```

```python
text
```

```python
# removes square brackets and extra spaces
text = re.sub(r'\[[0-9]*\]', ' ', text)  
text = re.sub(r'\s+', ' ', text)  

text
```

## Tokenizing Sentences

```python
# sentence tokenization
sentence_list = nltk.sent_tokenize(text)

sentence_list
```

```python
#
len(sentence_list)
```

```python
# question = 'what size were the WordPiece embeddings?'
# question = 'what batch size was used?'
question = 'how many epochs were used?'

# function to test if something is a noun
is_noun = lambda pos: pos[:2] == 'NN'

# do the nlp stuff
question_tokenized = nltk.word_tokenize(question)

question_nouns = [word for (word, pos) in nltk.pos_tag(question_tokenized) if is_noun(pos)] 

print(question_nouns)
```

```python
# sort capitals out
# look for words that are not nouns or verbs

matches = [s for s in sentence_list if any(xs in s for xs in question_nouns)]

matches
```

```python
len(matches)
```

```python
# rejoins the filtered sentences
filtered_sentences = ' '.join(matches)

filtered_sentences
```

```python
import requests
import json

url = "http://127.0.0.1:5000/api/run_pred/"

text = json.dumps({

    "q3": {
        
#         "context": filtered_sentences,
        "context": text,
        
        "question": question

    },

})

headers = {'content-type': 'application/json', 'Accept-Charset': 'UTF-8'}
r = requests.post(url, data=text, headers=headers)
print(r.text)
```
